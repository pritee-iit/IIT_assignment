```python
import numpy as np

data = np.array([[22.5, 19.0, 31.2, 28.7, 25.1],
                 [17.3, 22.8, 30.5, 26.4, 21.9],
                 [33.1, 29.6, 18.4, 24.0, 27.8],
                 [20.2, 23.5, 31.9, 28.1, 22.6]])

# Task 1 : Print shape and mean temperature per station :
print("Shape of the data array:", data.shape)
mean_per_station = np.mean(data, axis=1)
print("Mean temperature per station:",mean_per_station)
```

```python
# Task 2: Extract all records above 28 using boolean masking
above_28 = data[data > 28]
print("Temperature records above 28°C:", above_28)
```

```python
# Task 3 : Normaize to range [0,1]
normalized = (data - data.min()) / (data.max() - data.min())
print("Normaize Data:\n",np.round(normalized,2))
```
