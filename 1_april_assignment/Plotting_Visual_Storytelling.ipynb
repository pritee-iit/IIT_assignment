#### Visualizing Sales and Distribution Data Using Matplotlib and Seaborn

```python
import pandas as pd
import numpy as np

np.random.seed(42)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [12000, 15000, 13500, 17000, 16000, 19000,
         18500, 21000, 20000, 22500, 24000, 26000]

subjects = ['Math', 'Science', 'English', 'History', 'Art']
scores   = [78, 85, 72, 90, 65]

study_hours = np.random.uniform(1, 10, 100)
exam_scores = study_hours * 8 + np.random.normal(0, 5, 100)

product_sales = np.random.normal(loc=50, scale=10, size=200)
```

##### Task 1 — Line Plot: Monthly Sales Trend

```python
import matplotlib.pyplot as plt

# create plot
plt.figure(figsize=(10,5))
# Customize the line color, add markers, and include a title and axis labels.
plt.plot(months,sales,color="green", marker="o", linestyle="-",linewidth=2)

plt.title("Monthly Sales Tread",fontsize=14)
plt.xlabel("Month",fontsize=12)
plt.ylabel("Sales",fontsize=12)
plt.grid(True)

plt.show()
```

##### Task 2 — Bar Plot: Subject Score Comparison

```python
plt.figure(figsize=(10,5))
bars = plt.bar(subjects,scores,color="green",edgecolor="black")

# Add total value on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1,f'{height}',ha="center",va="bottom",fontsize=10)


plt.title("Score by Subject",fontsize=14)
plt.xlabel("Subjects",fontsize=12)
plt.ylabel("Scores",fontsize=12)
plt.ylim(0,100)
plt.grid(axis='y',linestyle="--",alpha=0.7)

plt.show()
```

##### Task 3 — Scatter Plot: Study Hours vs Exam Score

```python
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.figure(figsize=(10,5))

sns.scatterplot(x=study_hours,y=exam_scores,color="green",edgecolor="black")
plt.title("Study hours vs Exam score ",fontsize=14)
plt.xlabel("Subjects",fontsize=12)
plt.ylabel("Scores",fontsize=12)

plt.show()
```

##### Task 4 — Box Plot: Product Sales Distribution

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10,5))

sns.boxplot(y=product_sales,color="green")

plt.title("Product Sales Distribution",fontsize=14)
plt.ylabel("Sales",fontsize=12)

plt.show()
```
