#### SQL Basics — Building a Shipping Records Database

```python
import sqlite3 
```

```python
# Connect database
con = sqlite3.connect(':memory:')
cursor = con.cursor()
```

##### Task 1 — Create and Populate

```python
# Create and populate the table
cursor.execute('''
    CREATE TABLE shipping_mode_dimension(
               ship_mode TEXT,
               vehicle_company TEXT PRIMARY KEY,
               toll_required INTEGER
               )
''' 
)

data = [
    ('Delivery Truck','Ashok Leyland',1),
    ('Air Freight','Air India',0),
    ('Rail Cargo','Indian Railways',1),
    ('Delivery Truck','Tata Motors',1),
    ('Air Freight','SpiceJet Cargo',0),
    ('Rail Cargo','IRCTC Logistics',0)
]

cursor.executemany('''
INSERT INTO shipping_mode_dimension (ship_mode,vehicle_company,toll_required) VALUES (?,?,?)
''',data
)

# Commit
con.commit()
```

##### Task 2 — Query and Filter

```python
# (a) retrieve all rows where toll_required = 1
cursor.execute('SELECT * FROM shipping_mode_dimension WHERE toll_required = 1')
query_a = cursor.fetchall()
print("Rows with toll_required = 1")
for i in query_a:
    print(i)

# (b) count the total number of records in the table
cursor.execute('SELECT count(*) FROM shipping_mode_dimension')
query_b = cursor.fetchone()[0]
print(f"\nTotal records are :{query_b}")

# (c) retrieve only ship_mode and vehicle_company for all rows where ship_mode = 'Air Freight'.
cursor.execute('SELECT ship_mode,vehicle_company FROM shipping_mode_dimension where ship_mode = "Air Freight"')
query_c = cursor.fetchall()
print("\nRows with ship_mode = 'Air Freight'")
for k in query_c:
    print(k)
```

##### Task 3 — Modify and Clean

```python
# (a) Use ALTER TABLE to add a new column toll_amount of type INTEGER. 
cursor.execute("ALTER TABLE shipping_mode_dimension ADD COLUMN toll_amount INTEGER") 

# (b) Use UPDATE to set toll_amount = 200 where toll_required = 1 and toll_amount = 0 where toll_required = 0. 
cursor.execute("UPDATE shipping_mode_dimension SET toll_amount = 200 where toll_required = 1")  
cursor.execute("UPDATE shipping_mode_dimension SET toll_amount = 0 where toll_required = 0") 

# (c) Use DELETE to remove all rows where vehicle_company = 'Air India'
cursor.execute("DELETE FROM shipping_mode_dimension where vehicle_company = 'Air India'") 

# Commit
con.commit()

# Print result
cursor.execute("SELECT * FROM shipping_mode_dimension")
upadted_rows = cursor.fetchall()
 
print("\nRows after update & deletion")
for m in upadted_rows:
    print(m)

# Close connection
con.close()
```
