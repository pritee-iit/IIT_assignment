```python
# Student Performance Analytics System
# TASK 1: Grade Calculation Functions
# ----------------------------------
# return vs print:
# - 'return' sends a value back to the caller; can be stored in variables
# - 'print' displays a value on screen; does not allow further use

def calculate_average(subject1, subject2, subject3):
    # Calculates the average of three subject scores.
    return (subject1 + subject2 + subject3) / 3

def assign_grade(average_score):
    # Assigns a letter grade based on the average score.
    if 90 <= average_score <= 100:
        return 'A'
    elif 80 <= average_score < 90:
        return 'B'
    elif 70 <= average_score < 80:
        return 'C'
    elif 60 <= average_score < 70:
        return 'D'
    else:
        return 'F'

# Test Task 1
students = [
    {"Math": 85, "Science": 92, "English": 78},
    {"Math": 72, "Science": 68, "English": 75},
    {"Math": 95, "Science": 88, "English": 91}
]

for i, student in enumerate(students, start=1):
    avg = calculate_average(student["Math"], student["Science"], student["English"])
    grade = assign_grade(avg)
    print(f"Student {i}: Average = {avg:.2f}, Grade = {grade}")

# TASK 2: String Analysis Function with Default Parameters
# Default parameters:
# - Used when the caller does not provide that argument
# - Must come after required parameters in function definition

def count_vowels(text, case_sensitive=False):
    # Counts vowels in a string. If case_sensitive=False, counts all vowels ignoring case.
    vowels = 'aeiou'
    count = 0
    for char in text:
        if not case_sensitive:
            if char.lower() in vowels:
                count += 1
        else:
            if char in vowels:
                count += 1
    return count

# Test Task 2
texts = ["Hello World", "Education is Important", "PYTHON Programming"]

for txt in texts:
    count_default = count_vowels(txt)
    count_sensitive = count_vowels(txt, case_sensitive=True)
    print(f"Text: '{txt}' - Vowels (case_sensitive=False): {count_default}")
    print(f"Text: '{txt}' - Vowels (case_sensitive=True): {count_sensitive}")

# TASK 3: Dictionary Processing Function with Local Variables
# Local variables:
# - Variables defined inside a function
# - Not accessible outside the function

def find_top_performer(student_scores):
    # Finds and returns the name of the student with the highest score.
    highest_score = 0
    top_student = ""
    for student, score in student_scores.items():
        if score > highest_score:
            highest_score = score
            top_student = student
    return top_student, highest_score

# Test Task 3
scores = {
    "Rahul": 85,
    "Priya": 92,
    "Amit": 78,
    "Sneha": 95,
    "Karan": 88
}

top_student, top_score = find_top_performer(scores)
print(f"Top Performer: {top_student} with score {top_score}")
```
