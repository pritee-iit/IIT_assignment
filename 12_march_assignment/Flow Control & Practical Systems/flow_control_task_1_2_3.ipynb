```python
# Bookstore Inventory Management & Data Analysis
# Task 1: Inventory Restocking Analysis

inventory = {
    "Python Basics": 15,
    "Data Science": 8,
    "Web Development": 12,
    "Machine Learning": 5,
    "Database Design": 3,
    "Cloud Computing": 18
}

reorder_list = {}

print("=== Task 1: Inventory Restocking Analysis ===")
for book, stock in inventory.items():
    if stock < 10:
        quantity_needed = 10 - stock
        print(f"Reorder {book}: Need {quantity_needed} more books")
        reorder_list[book] = quantity_needed
    else:
        print(f"{book}: Stock sufficient")

print("\nBooks needing reorder:", reorder_list)

# Task 2: Nested Conditional Logic for Pricing

prices = {
    "Python Basics": 450,
    "Data Science": 850,
    "Web Development": 550,
    "Machine Learning": 950,
    "Database Design": 650,
    "Cloud Computing": 750
}

print("\n=== Task 2: Pricing & Stock Categorization ===")
for book, stock in inventory.items():
    price = prices[book]
    
    if stock < 10:
        # Book needs restock
        if price > 700:
            print(f"{book}: High-value restock needed")
        else:
            print(f"{book}: Standard restock needed")
    else:
        # Stock sufficient
        if price > 700:
            print(f"{book}: Premium item, stock OK")
        else:
            print(f"{book}: Standard item, stock OK")

# Task 3: Dictionary Data Cleaning

new_inventory = {
    "AI Fundamentals": 12,
    "Cybersecurity": None,
    "Blockchain": 8,
    "IoT Basics": None,
    "DevOps": 15
}

print("\n=== Task 3: Cleaned Inventory ===")
for book, stock in new_inventory.items():
    # Check for None - missing value
    if stock is None:
        stock = 0  # Replace None with 0
        new_inventory[book] = stock  # Update dictionary
    print(f"{book}: {stock}")
```
