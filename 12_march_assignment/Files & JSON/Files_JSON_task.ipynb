```python
# =========================
# Student Records Management
# =========================

# TASK 1: Text File Operations with Student Data
# 'r' = read mode
# 'w' = write mode
# 'a' = append mode

def write_attendance(filename, student_records):
    # Writes student attendance records to a text file.
    # Each record is a string: "StudentName,Present/Absent,Date"
    
    with open(filename, 'w') as file:
        for record in student_records:
            file.write(record + '\n')  # write each record on a new line

def read_attendance(filename):
    # Reads attendance records from a text file.
    # Returns a list of strings with newline characters removed.
    
    with open(filename, 'r') as file:
        lines = [line.strip() for line in file.readlines()]
    return lines

# Test Task 1
attendance_data = [
    "Rahul,Present,2024-01-15",
    "Priya,Absent,2024-01-15",
    "Amit,Present,2024-01-15",
    "Sneha,Present,2024-01-15",
    "Karan,Absent,2024-01-15"
]

write_attendance("attendance.txt", attendance_data)
print("Attendance Records:")
for i, line in enumerate(read_attendance("attendance.txt"), start=1):
    print(f"{i}. {line}")

# TASK 2: JSON File Operations with Student Grades

import json

def save_grades_to_json(filename, grades_dict):
    # Saves student grades dictionary to a JSON file.
    with open(filename, 'w') as file:
        json.dump(grades_dict, file, indent=4)

def load_grades_from_json(filename):
    # Loads student grades dictionary from a JSON file.
    with open(filename, 'r') as file:
        data = json.load(file)
    return data

# Test Task 2
student_grades = {
    "Rahul": {"Math": 85, "Science": 90, "English": 78},
    "Priya": {"Math": 92, "Science": 88, "English": 95},
    "Amit": {"Math": 76, "Science": 82, "English": 80}
}

save_grades_to_json("grades.json", student_grades)
grades_loaded = load_grades_from_json("grades.json")

print("\nStudent Average Scores:")
for student, subjects in grades_loaded.items():
    average = sum(subjects.values()) / len(subjects)
    print(f"{student}: {average:.2f}")

# TASK 3: Append Mode and Data Processing

def add_attendance_record(filename, new_record):
    # Appends a single new attendance record to the file.
    with open(filename, 'a') as file:
        file.write(new_record + '\n')

def count_present_students(filename):
    # Counts how many students are marked 'Present' in the attendance file.
    records = read_attendance(filename)
    count = sum(1 for record in records if record.split(',')[1].strip() == "Present")
    return count

# Test Task 3
add_attendance_record("attendance.txt", "Neha,Present,2024-01-15")
add_attendance_record("attendance.txt", "Vikram,Absent,2024-01-15")

total_present = count_present_students("attendance.txt")
print(f"\nTotal students present: {total_present}")
```
