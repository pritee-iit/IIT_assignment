```python
# E-Commerce Order Processing System
# TASK 1: Order Invoice with Line Numbers
# Use enumerate with start=1 so invoice line numbers start from 1

order_items = ["Laptop", "Mouse", "Keyboard", "Monitor", "USB Cable"]
prices = [45000, 500, 1500, 12000, 300]

invoice_lines = []
total_amount = 0

print("===== Invoice =====")
for line_num, item in enumerate(order_items, 1):
    price = prices[line_num - 1]
    print(f"Line {line_num}: {item} - Rs. {price}")
    invoice_lines.append(f"{line_num}. {item}")
    total_amount += price

print(f"Total Amount: Rs. {total_amount}")
print("Invoice Lines List:", invoice_lines)

# TASK 2: Position-Based Discount System
# Discounts based on position: even indices = 10%, odd indices = 5%

products = ["Smartphone", "Headphones", "Charger", "Case", "Screen Guard", "Power Bank"]
base_prices = [25000, 2000, 800, 500, 300, 1500]

final_prices = []

print("\n===== Discounted Prices =====")
for index, product in enumerate(products):
    original_price = base_prices[index]
    if index % 2 == 0:
        discount = 0.10
    else:
        discount = 0.05
    savings = original_price * discount
    discounted_price = original_price - savings
    print(f"Position {index}: {product} - Original: Rs.{original_price}, Discounted: Rs.{discounted_price}, You save: Rs.{savings}")
    final_prices.append(discounted_price)


# TASK 3: Priority Queue with Index Filtering
# Use enumerate to assign priority based on index
tickets = ["Password Reset", "Bug Report", "Feature Request", "Account Locked", "Payment Issue", "General Query", "Refund Request"]

high_priority_count = 0

print("\n===== Ticket Priority =====")
for index, ticket in enumerate(tickets):
    if index % 3 == 0:
        priority = "High Priority"
        high_priority_count += 1
    elif index <= 2:
        priority = "Medium Priority"
    else:
        priority = "Low Priority"
    print(f"Ticket #{index}: {ticket} - {priority}")

print(f"Total High Priority Tickets: {high_priority_count}")
```
