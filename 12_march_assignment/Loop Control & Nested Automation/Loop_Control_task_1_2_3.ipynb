```python
# Student Grade Processing & Tournament Report
# Task 1: Student Average Grades

school_data = [
    ["Alice", 85, 92, 78],
    ["Bob", 72, 68, 81],
    ["Charlie", 95, 88, 91],
    ["Diana", 63, 59, 67],
    ["Eve", 88, 91, 85]
]

print("=== Task 1: Student Average Grades ===")
for student in school_data:
    name = student[0]
    total = 0
    for score in student[1:]:
        total += score
    average = total / (len(student) - 1)
    
    if average < 70:
        continue
    print(f"{name}: Average = {average:.2f} - PASS")



# Task 2: Tournament Matches
tournament = [
    ["Player_A1", "Player_A2", "Player_A3"],
    ["Player_B1", "Player_B2", "Player_B3"],
    ["Player_I1", "Player_I2", "Player_I3"]
]

print("\n=== Task 2: Tournament Matches ===")
special_found = False

for group in tournament:
    for i in range(len(group)-1):
        player1 = group[i]
        player2 = group[i+1]
        print(f"Match: {player1} vs {player2}")
        
        if "I2" in player1 or "I2" in player2:
            special_found = True
            break
    if special_found:
        break

if special_found:
    print("Tournament paused - Special player found")

# Task 3: Number Pattern (Fizz/Buzz)
print("\n=== Task 3: Number Pattern (Fizz/Buzz) ===")
number = 1
while number <= 20:
    if number % 3 == 0 and number % 5 == 0:
        number += 1
        continue
    if number % 3 == 0:
        print(f"{number}: Fizz")
    elif number % 5 == 0:
        print(f"{number}: Buzz")
    else:
        print(number)
    number += 1
```
