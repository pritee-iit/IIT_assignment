#### Public REST API

```python
import pandas as pd
import requests
API_url = "https://restcountries.com/v3.1/all?fields=name,population,region,area"
```

Task 1 — Fetch and Validate the Response

```python
try:
    response = requests.get(API_url)
    print("Response status code :",response.status_code)

    if response.status_code == 200:
        get_data = response.json()
        print(f"Total number of countries retrieved: {len(get_data)}")
    else:
        print(f"Error: Response status code :",response.status_code)
except requests.exceptions.RequestException as e:
    print(f"Failed:{e}")
```

Task 2 — Extract and Display Key Fields

```python
rows = []

for country in get_data:
    row = {
        "common_name":country.get("name",{}).get("common",None),
        "population":country.get("population",None),
        "region":country.get("region",None),
        "area":country.get("area",None)
    }
    rows.append(row)

# create datafram
df = pd.DataFrame(rows)

# Display 10 rows from df
print("\n First 10 rows:")
print(df.head(10))
```

Task 3 — Summarize the Data

```python
# Region with highest total population
population_by_region = df.groupby('region')["population"].sum()
top_region = population_by_region.idxmax()
top_region_val = population_by_region.max()

# Counry with largest area
if df["area"].notna().any():
    largest_country_area = df.loc[df["area"].idxmax()]
    largest_country_name = largest_country_area["common_name"]
    largest_country_area_val = largest_country_area["area"]
else:
    largest_country_area = None
    largest_country_name = None
    largest_country_area_val = None


# Final Summery
print("\nFinal Summery:")
print(f"Region with highest population: {top_region} ({top_region_val:,})")
print(f"Country with largest area :{largest_country_name} ({largest_country_area_val:,}) km")
```
