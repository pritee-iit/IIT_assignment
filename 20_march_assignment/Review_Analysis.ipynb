```python
import pandas as pd
import random
import matplotlib.pyplot as plt

# Set the seed for reproducibility
random.seed(42)

# Generate 100 product reviews
categories = ['Electronics', 'Clothing', 'Books', 'Sports']
cities = ['Mumbai', 'Delhi', 'Bangalore', 'Chennai']

# Create synthetic data
data = []
for i in range(100):
    data.append({
        'product': random.choice(['Laptop', 'T-Shirt', 'Novel', 'Shoes', 'Phone']),
        'category': random.choice(categories),
        'rating': random.randint(1, 5),
        'price': random.randint(500, 20000),
        'city': random.choice(cities)
    })

# Create DataFrame
df = pd.DataFrame(data)
```

Task - 1 Preview and Explore

```python
first_10_rows = df.head(10)
last_5_rows = df.tail(5)
dataset_shape = df.shape
dataset_info = df.info()
summary_stats = df.describe()
```

```python
print(first_10_rows)
print(last_5_rows)
print(dataset_shape)
print(dataset_info)
print(summary_stats)
```

Task 2 - Filter with Conditions

```python
# Filter all reviews from Bangalore
bangalore_df = df[df['city'] == 'Bangalore']

# Filter products with 5-star rating AND price > 10000
high_rating_expensive = df[(df['rating'] == 5) & (df['price'] > 10000)]

# Filter Electronics OR Books with rating < 3
low_rating_electronics_books = df[((df['category'] == 'Electronics') | (df['category'] == 'Books')) & (df['rating'] < 3)]
```

Task 3 - Value Counts and Visualization

```python
# Count reviews per category
category_counts = df['category'].value_counts()

# Find city with most reviews
most_reviews_city = df['city'].value_counts().idxmax()

# Create visualizations
category_bar_chart = df['category'].value_counts().plot(kind='bar', title='Review Count by Category')
rating_histogram = df['rating'].plot(kind='hist', bins=5, title='Rating Distribution')

# Show the plots
plt.show()

print(category_counts)
print(most_reviews_city)
```
