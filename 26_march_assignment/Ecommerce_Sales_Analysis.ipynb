Data Loading

```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
```

Task 1 - Basic Grouping and Single Aggregations

```python
# 1. Total sales amount for each region
region_sales = df.groupby('region')['sales_amount'].sum().reset_index()
region_sales = region_sales.sort_values('sales_amount', ascending=False)

# 2. Count the number of transactions for each product category
category_transactions = df.groupby('category')['transaction_id'].count().reset_index()

# 3. Average sales amount per salesperson
salesperson_avg_sales = df.groupby('salesperson')['sales_amount'].mean().reset_index()

# Display Results
print("Total Sales per Region:")
print(region_sales.head())

print("\nNumber of Transactions per Category:")
print(category_transactions.head())

print("\nAverage Sales per Salesperson:")
print(salesperson_avg_sales.head())
```

Task 2 - Multi-Column Grouping and Multiple Aggregations

```python
# 1. Group by region and category to calculate total sales for each combination
region_category_sales = df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()

# 2. Multi-metric aggregation for each salesperson
salesperson_stats = df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count']).reset_index()

# 3. Sort salesperson by total sales in descending order
salesperson_stats_sorted = salesperson_stats.sort_values('sum', ascending=False)

# 4. Find the category with maximum total revenue
top_category = region_category_sales.groupby('category')['sales_amount'].sum().idxmax()

# Display Results
print("\nTotal Sales by Region and Category:")
print(region_category_sales.head())

print("\nSalesperson Stats (Total Sales, Average, Count):")
print(salesperson_stats_sorted.head())

print(f"\nCategory with Maximum Total Revenue: {top_category}")
```

Task 3 - Custom Aggregation and Complete Sales Report

```python
# 1. Define a custom aggregation function for sales range
def sales_range(x):
    return x.max() - x.min()

# 2. Apply custom aggregation along with standard aggregations by region
region_sales_analysis = df.groupby('region')['sales_amount'].agg(['sum', 'mean', 'min', 'max', sales_range]).reset_index()

# 3. Create a final summary report by region
region_summary = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()

# Display Results
print("\nSales Analysis by Region (Including Custom Sales Range):")
print(region_sales_analysis.head())

print("\nFinal Sales Summary by Region:")
print(region_summary.head())
```

##### Explanation of Multi-level Columns:

<li/>In Task 3, when we use a dictionary for aggregation, the resulting columns are multi-level. 

<li/>For example, when we apply: <br/>
df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()

<li/>DataFrame will have a two-level column structure. The first level contains the names of the fields ('sales_amount', 'customer_id'), and the second level contains the aggregation functions applied to them ('sum', 'mean', 'count'). This structure helps keep track of multiple aggregations on the same field.
