Task 1 - Create the Pandas DataFrame

```python
import pandas as pd

# Data for students
data = {
    'StudentID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Ethan', 'Fiona', 'George', 'Hannah', 'Ivan', 'Julia'],
    'Age': [20, 21, 19, 22, 20, 21, 23, 19, 22, 20],
    'Marks': [88, 75, 92, 65, 78, 85, 70, 95, 60, 80],
    'Grade': ['A', 'B', 'A', 'C', 'B', 'A', 'B', 'A', 'C', 'B']
}

# Create DataFrame
students_df = pd.DataFrame(data)

# Display the DataFrame
students_df
```

Task 2 - Save the DataFrame as a CSV file

```python
# Save DataFrame as CSV file
students_df.to_csv('students.csv', index=False)
```

Above code saves the DataFrame students_df to a CSV file(students.csv), index = false used for exclude the index column.

Task 3 - Reload the CSV file into a new DataFrame

```python
# Reload the CSV file into a new DataFrame
df_reload = pd.read_csv('students.csv')

# Display the DataFrame to verify
df_reload
```

ABove code reload the csv data into new dataframe

Task 6 - Explore the DataFrame using .shape, .dtypes, and .info()

```python
# Display the shape of the DataFrame
df_reload.shape

# Display the data types of each column
df_reload.dtypes

# Display detailed information about the DataFrame
df_reload.info()
```

1) shape returns the number of rows and columns.

2) dtypes shows the data types of each column.

3) info() provides more detailed information, such as the number of non-null entries in each column.
