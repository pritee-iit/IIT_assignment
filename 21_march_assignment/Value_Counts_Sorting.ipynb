```python
import pandas as pd

# Sales Data
sales_data = {
    'order_id': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110, 111, 112],
    'customer': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob', 'Alice', 'David', 'Charlie', 'Bob', 'Alice', 'David', 'Charlie'],
    'product': ['Laptop', 'Mouse', 'Laptop', 'Keyboard', 'Mouse', 'Monitor', 'Laptop', 'Mouse', 'Keyboard', 'Laptop', 'Monitor', 'Keyboard'],
    'category': ['Electronics', 'Electronics', 'Electronics', 'Electronics', 'Electronics', 'Electronics', 'Electronics', 'Electronics', 'Electronics', 'Electronics', 'Electronics', 'Electronics'],
    'amount': [50000, 500, 50000, 1500, 500, 12000, 50000, 500, 1500, 50000, 12000, 1500],
    'city': ['Mumbai', 'Delhi', 'Mumbai', 'Bangalore', 'Delhi', 'Mumbai', 'Delhi', 'Bangalore', 'Mumbai', 'Delhi', 'Bangalore', 'Mumbai']
}

df = pd.DataFrame(sales_data)
```

Task 1: Value Counts and Sorting

```python
# Part A
customer_order_count = df['customer'].value_counts().sort_values(ascending=False)
print(customer_order_count)
```

```python
# Part B
top_3_products = df['product'].value_counts().head(3)
print(top_3_products)
```

value_counts() counts the occurrences of unique values in a column. It is useful for finding frequency distributions.
e.g. df['customer'].value_counts() will tell us how many times each customer has placed an order.

sort_values(ascending=False): This method sorts the data in ascending or descending order. 
ascending=False sorts the data in descending order (i.e., from largest to smallest).


Difference between ascending=True/ascending=False:

ascending=True sorts the data in increasing order.

ascending=False sorts the data in decreasing order.

Task 2 - Handling Duplicates

```python
# Part A
df_no_duplicates = df.drop_duplicates(subset=['customer', 'product'], keep='first')
print(df_no_duplicates)
```

```python
# Part B
unique_cities = df['city'].unique()
print(unique_cities)
```

drop_duplicates() removes duplicate rows based on specified columns.
subset parameter specifies the columns to consider when identifying duplicates.


keep='first': This keeps the first occurrence of a duplicate row and removes the subsequent ones.
inplace=True: If set to True, the method modifies the DataFrame in place, rather than returning a new one.
If False, the method returns a new DataFrame without altering the original.

Difference between keep='first/keep='last':
keep='first': Keeps the first occurrence and removes duplicates after that.
keep='last': Keeps the last occurrence and removes duplicates before it.

unique(): This method returns an array of the unique values from a column.

Task 3 - Group By Operations

```python
# Part A
customer_spending = df.groupby('customer')['amount'].sum().sort_values(ascending=False)
print(customer_spending)
```

```python
# Part B
city_sales = df.groupby('city')['amount'].sum()
print(city_sales)
```

```python
# Part C
customer_product_count = df.groupby('customer')['product'].nunique()
print(customer_product_count)
```

groupby() is used to split the data into groups based on the values of one or more columns

Aggregation functions:

sum() calculates the total value for a numerical column in each group.
In Part A, we use sum() to calculate the total amount spent by each customer.

nunique() counts the number of unique values in a column within each group.
In Part C, we use nunique() to count how many different products each customer purchased.

Sorting:
We use sort_values(ascending=False) to sort the total amount spent by each customer in descending order to find the top spender.

Grouping by multiple columns:
can group by multiple columns by passing a list of columns to groupby(). 
e.g. df.groupby(['customer', 'city']) would group the data by both customer and city.
